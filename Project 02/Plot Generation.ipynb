```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.dpi'] = 100
```

```python
x = [1, 2, 3, 4, 5]
y1 = [0.012, 0.007, 0.007, 0.006, 0.007]
y2 = [0.012, 0.010, 0.007, 0.007, 0.007]
plt.figure()
plt.plot(x, y1, "r")
plt.plot(x, y2, "b")
plt.legend(["TCP", "TLS"])
plt.title("Execution Time For Different Execution Trials")
plt.xlabel("Number of execution")
plt.ylabel("Execution time")
plt.grid(True)
plt.savefig("plot.png")
```
